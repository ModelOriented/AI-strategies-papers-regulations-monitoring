```python
import pandas as pd
```

```python
from glob import glob
import plotly.express as px
```

```python
files_to_read = list(glob("../../mars/s2orc/metadata/ai/*.jsonl.gz"))
```

```python
dfs = [pd.read_json(file, lines=True, compression=None) for file in files_to_read]
```

```python
df = pd.concat(dfs)
```

```python
df['in_citations_count'] = df['inbound_citations'].str.len()
df.describe()
```

```python
px.histogram(df['in_citations_count'], log_y=True)
```

```python
df2 = df[df['in_citations_count']!=0]
```
