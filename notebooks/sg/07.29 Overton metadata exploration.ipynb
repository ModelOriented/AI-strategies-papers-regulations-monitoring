```python
import spacy

nlp = spacy.load('en_core_web_sm')

title_docs = df3['Title'].apply(nlp)
```

```python
df3['Translated title'] = df3['Translated title'].fillna('')
translated_title_docs = df3['Translated title'].apply(nlp)
```

```python
px.histogram(title_docs.str.len())
```

```python
title_docs[title_docs.str.len()==1]
```

```python
ai_patterns = ['AI', 'artificial intelligence', 'machine learning']
```

```python
from spacy.matcher import PhraseMatcher


matcher = PhraseMatcher(nlp.vocab, attr="NORM")
for pattern in ai_patterns:
    matcher.add('AI', [nlp(pattern)])
```

```python
result = title_docs.apply(matcher)
results_translated = translated_title_docs.apply(matcher)
```

```python
sum((result.str.len()!=0) | (results_translated.str.len()!=0))
```

```python
df_ai = df3[(result.str.len()!=0) | (results_translated.str.len()!=0)]
```

```python
df_ai['year']= df_ai['date'].dt.year

```
